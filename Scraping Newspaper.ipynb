```python
from newspaper import Article

# masukkan url
url1 = 'http://suryamalang.tribunnews.com/2018/07/04/sukses-raih-6-poin-di-piala-aff-19-timnas-indonesia-u-19-jangan-keburu-puas-dulu'

'''ini cuma buat nambahin ke url aja biar beritanya ga kepotong
soalnya kalo berita tribunnews ada per part halaman yang nunjukin beritanya
jadi di url ditambahin ?page=all biar nunjukin semua tulisan di konten beritanya'''
if "tribunnews" in url1 :
    url1 = url1 + '?page=all'

a1 = Article(url1, language='id') #language disesuaikan dengan bahasa di konten beritanya
a1.download() #download file htmlnya
a1.parse() #parsing ke xml
a1.nlp()

print("\n=========Judul Berita========\n")
print(a1.title)
print("\n=========Penulis Berita========\n")
print(a1.authors)
print("\n=========Tanggal Berita========\n")
print(a1.publish_date)
print("\n=========Gambar Berita========\n")
print(a1.top_image)
print("\n=========Video Berita========\n")
print(a1.movies)
print("\n=========Teks Konten Berita========\n")
print(a1.text)

#ga semua data di atas ada. Kadang ada yang kosong. Kadang textnya ga kedetect.

print("\n=========Kata Kunci Berita========\n")
print(a1.keywords)
print("\n=========Ringkasan Berita========\n")
print(a1.summary)
```
